```python
import matplotlib.pyplot as plt
import csv
#==================
def mix_plot(title_a,title_b,savename,lista,listb,listc,va1,va2,y1min=40,y1max=95,y2min=20,y2max=110,xmax=20):
#==================   
    name_list=['train_org','valid_org','train_dbscan','valid_dbscan','train_autoencoder','valid_autoencoder']
    #transfer to %
    acc_list=[[],[],[]]
    val_acc_list=[[],[],[]]   
    loss_list=[[],[],[]]
    val_loss_list=[[],[],[]]
    for i in lista['acc']:
        acc_list[0].append(round(i*100,1))
    for i in lista['val_acc']:
        val_acc_list[0].append(round(i*100,1))
    for i in lista['loss']:
        loss_list[0].append(round(i*100,1))
    for i in lista['val_loss']:
        val_loss_list[0].append(round(i*100,1))

    for i in listb['acc']:
        acc_list[1].append(round(i*100,1))
    for i in listb['val_acc']:
        val_acc_list[1].append(round(i*100,1))
    for i in listb['loss']:
        loss_list[1].append(round(i*100,1))
    for i in listb['val_loss']:
        val_loss_list[1].append(round(i*100,1))
        
    for i in listc['acc']:
        acc_list[2].append(round(i*100,1))
    for i in listc['val_acc']:
        val_acc_list[2].append(round(i*100,1))
    for i in listc['loss']:
        loss_list[2].append(round(i*100,1))
    for i in listc['val_loss']:
        val_loss_list[2].append(round(i*100,1))

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,5))
    plt.subplots_adjust( left= 0, bottom= 0, right=1, top= 1,hspace =0.1, wspace =0.1 )
    #fig.suptitle('A tale of 2 subplots')
    x_text=[]
    y_text=[]
    for i in range(3):
        a=len(acc_list[i])-1
        b=acc_list[i][a]
        x_text.append(a)
        y_text.append(b)
        a=len(val_acc_list[i])-1
        b=val_acc_list[i][a]
        x_text.append(a)
        y_text.append(b)        
        
    ax1.plot(acc_list[0],    color='red',  linestyle='solid')
    ax1.plot(val_acc_list[0],color='red',  linestyle='dashed')
    ax1.plot(acc_list[1],    color='green',linestyle='solid')
    ax1.plot(val_acc_list[1],color='green',linestyle='dashed')
    ax1.plot(acc_list[2],    color='blue', linestyle='solid')
    ax1.plot(val_acc_list[2],color='blue', linestyle='dashed')
    ax1.text(x_text[0],y_text[0], "%s" %y_text[0], ha="left",va=va1[0])    
    ax1.text(x_text[1],y_text[1], "%s" %y_text[1], ha="left",va=va1[1])
    ax1.text(x_text[2],y_text[2], "%s" %y_text[2], ha="left",va=va1[2])    
    ax1.text(x_text[3],y_text[3], "%s" %y_text[3], ha="left",va=va1[3])    
    ax1.text(x_text[4],y_text[4], "%s" %y_text[4], ha="left",va=va1[4])    
    ax1.text(x_text[5],y_text[5], "%s" %y_text[5], ha="left",va=va1[5])    
    
    ax1.set_title(title_a,fontweight="bold", size=20)
    ax1.set_ylabel('accuracy %',fontsize=12)
    ax1.set_xlabel('epoch',fontsize=12)
    ax1.legend(name_list, loc='lower right')
    ax1.set_ylim([y1min,y1max])
    ax1.autoscalex_on=0
    ax1.set_yticks(list(range(y1min,y1max,10)))
    ax1.set_xticks(list(range(0,xmax)))
#    ax1.set_xticklabels(list(range(1,len(mhistory.history['acc'])+1,xtick_index)))
    ax1.grid()
   
    x_text=[]
    y_text=[]
    for i in range(3):
        a=len(loss_list[i])-1
        b=loss_list[i][a]
        x_text.append(a)
        y_text.append(b)
        a=len(val_loss_list[i])-1
        b=val_loss_list[i][a]
        x_text.append(a)
        y_text.append(b)     
    ax2.plot(loss_list[0]    ,color='red'  ,  linestyle='solid')
    ax2.plot(val_loss_list[0],color='red'  ,  linestyle='dashed')    
    ax2.plot(loss_list[1]    ,color='green',  linestyle='solid')
    ax2.plot(val_loss_list[1],color='green',  linestyle='dashed')    
    ax2.plot(loss_list[2]    ,color='blue' ,  linestyle='solid')    
    ax2.plot(val_loss_list[2],color='blue' ,  linestyle='dashed')
    ax2.text(x_text[0],y_text[0], "%s" %y_text[0], ha="left",va=va2[0])    
    ax2.text(x_text[1],y_text[1], "%s" %y_text[1], ha="left",va=va2[1])
    ax2.text(x_text[2],y_text[2], "%s" %y_text[2], ha="left",va=va2[2])    
    ax2.text(x_text[3],y_text[3], "%s" %y_text[3], ha="left",va=va2[3])    
    ax2.text(x_text[4],y_text[4], "%s" %y_text[4], ha="left",va=va2[4])    
    ax2.text(x_text[5],y_text[5], "%s" %y_text[5], ha="left",va=va2[5]) 
    ax2.set_title(title_b,fontweight="bold", size=20)
    ax2.set_ylabel('loss %',fontsize=12)
    ax2.set_xlabel('epoch',fontsize=13)
    ax2.legend(name_list, loc='upper right')
    ax2.set_ylim([y2min,y2max])
    ax2.autoscalex_on=0
    ax2.set_yticks(list(range(y2min,y2max,10)))
    ax2.set_xticks(list(range(0,xmax)))
#    ax2.set_xticks(list(range(0,len(mhistory.history['val_acc']),xtick_index)))
#    ax2.set_xticklabels(list(range(1,len(mhistory.history['val_acc'])+1,xtick_index)))
    ax2.grid()
    plt.show()
    fig.savefig('%s%s.jpg'%(path,savename), bbox_inches='tight', dpi=150)
    i=[]
    j=[]
    for k in range (3):
        i.append(acc_list[k][len(acc_list[k])-1])
        i.append(val_acc_list[k][len(val_acc_list[k])-1]) 
        j.append(loss_list[k][len(loss_list[k])-1])
        j.append(val_loss_list[k][len(val_loss_list[k])-1])
    a=['Name']+name_list
    i=['ACC']+i
    j=['LOSS']+j
    print(a)
    print(i)
    print(j)
    
    #write header1,header2 and rows to excel /csv
    with open('%s%s.csv'%(path,savename),'w', newline='') as csvfile:               
        writer = csv.writer(csvfile)
        writer.writerow(a)
        writer.writerow(i)
        writer.writerow(j)
    return   
```
